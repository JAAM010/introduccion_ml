## üõ†Ô∏è 1. ¬øQu√© es un entorno virtual?

Un entorno virtual es una herramienta que permite crear un espacio aislado en el sistema de archivos para un proyecto de Python. Esto significa que puedes tener diferentes versiones de paquetes y dependencias para diferentes proyectos sin que interfieran entre s√≠.

## üéØ 2. ¬øPor qu√© usar entornos virtuales?

Un entorno virtual te permite:
- Aislar las dependencias de cada proyecto.
- Evitar conflictos entre versiones de paquetes.
- Tener reproducibilidad entre equipos o servidores.

## üê≥ 9. Introducci√≥n a Docker para proyectos de ML/IA

üéØ ¬øPor qu√© usar Docker?

Docker permite crear contenedores: entornos ligeros y portables que incluyen todo lo necesario para ejecutar tu proyecto (Python, librer√≠as, modelo, dependencias, etc.).

Esto garantiza que:
- Tu modelo funcione igual en producci√≥n que en tu m√°quina.
- No haya conflictos entre entornos.
- Sea m√°s f√°cil escalar, migrar y automatizar.


‚öôÔ∏è ¬øQu√© contiene un contenedor Docker?

Un contenedor t√≠pico de ML puede incluir:
- Python y sus dependencias (numpy, scikit-learn, fastapi, etc.)
- Tu c√≥digo (src/, main.py)
- Archivos del modelo (model.pkl, model.joblib)
- Herramientas del sistema (Linux base, bash, curl, etc.)

‚∏ª

üìÑ Ejemplo de Dockerfile

```dockerfile
# Imagen base con Python
FROM python:3.11-slim

# Crear directorio de trabajo
WORKDIR /app

# Copiar dependencias y c√≥digo
COPY pyproject.toml poetry.lock ./
RUN pip install poetry && poetry install --no-root

COPY . .

# Comando por defecto al correr el contenedor
CMD ["poetry", "run", "python", "main.py"]
```
